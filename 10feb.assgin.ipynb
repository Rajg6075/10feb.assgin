```python
Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode
of file opening.
ANS= 

In Python, the function used to open a file is open(). It is used to open a file on the disk and returns a file object that allows you to read from or write to the file.

The open() function takes two arguments: the file path (the name of the file to be opened) and the mode in which the file should be opened.

The different modes of opening a file are as follows:

Read Mode ('r'):
This is the default mode. It opens the file for reading, and if the file does not exist, it raises a FileNotFoundError. The file pointer is placed at the beginning of the file, allowing you to read data from the file.

Write Mode ('w'):
This mode opens the file for writing. If the file already exists, it will truncate the file to zero length. If the file does not exist, it will create a new file. The file pointer is placed at the beginning of the file, allowing you to write data to the file. Be cautious, as using this mode will overwrite the existing content of the file.

Append Mode ('a'):
This mode opens the file for writing, but it appends data to the end of the file instead of truncating it. If the file does not exist, it will create a new file. The file pointer is placed at the end of the file, allowing you to append data to the file without overwriting the existing content.

```

```python
Q3. Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then
close the file. Open this file and read the content of the file.
ANS= 
def create_and_write_file(file_name, content):
    try:
        # Open the file in write mode
        with open(file_name, 'w') as file:
            # Write the content to the file
            file.write(content)
            print("File created and written successfully.")
    except IOError:
        print("Error: Unable to create or write to the file.")

def read_file(file_name):
    try:
        # Open the file in read mode
        with open(file_name, 'r') as file:
            # Read the content from the file
            content = file.read()
            print("Content of the file:")
            print(content)
    except IOError:
        print("Error: Unable to read the file.")

if __name__ == "__main__":
    file_name = "data_scientist.txt"
    content_to_write = "I want to become a Data Scientist"

    create_and_write_file(file_name, content_to_write)
    read_file(file_name)
```

```python
Q4. Explain the following with python code: read(), readline() and readlines().
Ans=
read(): This method is used to read the entire content of a file as a single string.

def read_file_with_read(file_name):
    try:
        # Open the file in read mode
        with open(file_name, 'r') as file:
            # Read the entire content of the file as a single string
            content = file.read()
            print("Content of the file (read() method):")
            print(content)
    except IOError:
        print("Error: Unable to read the file.")

if __name__ == "__main__":
    file_name = "example.txt"
    read_file_with_read(file_name)

    
readline(): This method is used to read a single line from the file at a time. Each time you call readline(), it reads the next line in the file.

def read_file_with_readline(file_name):
    try:
        # Open the file in read mode
        with open(file_name, 'r') as file:
            print("Content of the file (readline() method):")
            # Read each line from the file and print it
            line = file.readline()
            while line:
                print(line, end="")
                line = file.readline()
    except IOError:
        print("Error: Unable to read the file.")

if __name__ == "__main__":
    file_name = "example.txt"
    read_file_with_readline(file_name)

    
readlines(): This method is used to read all the lines from the file and store them as a list of strings. Each element in the list represents one line from the file.


def read_file_with_readlines(file_name):
    try:
        # Open the file in read mode
        with open(file_name, 'r') as file:
            # Read all lines from the file and store them as a list
            lines = file.readlines()
            print("Content of the file (readlines() method):")
            for line in lines:
                print(line, end="")
    except IOError:
        print("Error: Unable to read the file.")

if __name__ == "__main__":
    file_name = "example.txt"
    read_file_with_readlines(file_name)
```

```python
Q5. Explain why with statement is used with open(). What is the advantage of using with statement and
open() together?
ANS=

The with statement in Python is used in conjunction with the open() function to simplify the management of resources, especially when working with files. It ensures that resources like files are properly handled and automatically cleaned up after their usage, even in the presence of exceptions or errors.

The advantage of using the with statement with open() for file handling can be summarized as follows:

Automatic resource management: When using with, Python automatically takes care of opening the file and closing it when it's no longer needed. The with statement guarantees that the file is properly closed, regardless of how the block is exited, whether through normal execution or due to an exception. This helps prevent resource leaks and ensures that the file's resources are released promptly after its usage.

Clean and concise code: The with statement makes the code cleaner and more concise. It eliminates the need to explicitly open and close the file using file.open() and file.close(). Instead, the file is automatically opened and closed within the with block, making the code easier to read and maintain.

```

```python
Q6. Explain the write() and writelines() functions. Give a suitable example.
ANS=

Both write() and writelines() functions are used for writing data into a file in Python.

write(): This method is used to write a single string (or a bytes object) to a file. If the file already exists, it opens the file in write mode and writes the content at the current file position, overwriting the existing content. If the file does not exist, it creates a new file and writes the content into it.

writelines(): This method is used to write a list of strings (or bytes objects) to a file. Each element of the list represents a line to be written into the file. It does not add line breaks automatically, so you need to include them in the strings if you want to write each element on a new line.
```
